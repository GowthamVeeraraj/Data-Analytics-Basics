# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
age=np.random.randint(18,47,100)
age1=np.random.randint(15,40,100)
```

```python
age
```

```python
age1
```

```python
plt.figure(figsize=(16,9))
plt.hist([age,age1],bins=[15,20,25,30,35,40,45],rwidth=0.8,histtype="bar",color=['g','y'],orientation="vertical")
#plt.hist(age1,bins=[15,20,25,30,35,40,45],rwidth=0.8,histtype="bar",color='y')
```

```python
plt.hist(
    x,
    bins=None,
    range=None,
    density=False,
    weights=None,
    cumulative=False,
    bottom=None,
    histtype='bar',
    align='mid',
    orientation='vertical',
    rwidth=None,
    log=False,
    color=None,
    label=None,
    stacked=False,
    *,
    data=None,
    **kwargs,
)
Docstring:
Plot a histogram.

Compute and draw the histogram of *x*.  The return value is a tuple
(*n*, *bins*, *patches*) or ([*n0*, *n1*, ...], *bins*, [*patches0*,
*patches1*, ...]) if the input contains multiple data.  See the
documentation of the *weights* parameter to draw a histogram of
already-binned data.
```

```python
c=['A','B','C','D','E']
c1=[30,10,20,25,10]
c2=[40,15,17,19,39]
c3=[35,12,50,46,21]
```

```python
plt.bar(c,c1,width=0.5,color='c',align='edge',edgecolor='k',ls="--",alpha=0.8)
```

```python
w=np.arange(len(c))
width=0.2
plt.barh(w,c1,width,color='g',align='edge')
plt.barh(w+width,c2,width,color='y',align='edge')
plt.barh(w+(width*2),c3,width,color='b',align='edge')
plt.yticks(w+width,c)
plt.show
```

```python
import pandas as pd
df=pd.DataFrame([c1,c2,c3],columns=c)
df
```
